```python
import pandas as pd

# Load CSV files
citibike_may23_df = pd.read_csv('Resources/2023 May/202305-citibike-tripdata_1.csv', low_memory=False)
citibike_may24_df = pd.read_csv('Resources/2024 May/202405-citibike-tripdata_1.csv', low_memory=False)
```

```python
citibike_may23_df.info()
citibike_may24_df.info()
```

```python
citibike_may23_df = citibike_may23_df.dropna()
citibike_may24_df = citibike_may24_df.dropna()

citibike_may23_df.info()
citibike_may24_df.info()
```

```python
# Convert the started_at and ended_at columns to datetime
citibike_may23_df = citibike_may23_df.assign(
    started_at=pd.to_datetime(citibike_may23_df['started_at'], errors='coerce'),
    ended_at=pd.to_datetime(citibike_may23_df['ended_at'], errors='coerce')
)
citibike_may24_df = citibike_may24_df.assign(
    started_at=pd.to_datetime(citibike_may24_df['started_at'], errors='coerce'),
    ended_at=pd.to_datetime(citibike_may24_df['ended_at'], errors='coerce')
)

# Convert the start_station_id and end_station_id columns to numeric
citibike_may23_df.loc[:, 'start_station_id'] = pd.to_numeric(citibike_may23_df['start_station_id'], errors='coerce')
citibike_may23_df.loc[:, 'end_station_id'] = pd.to_numeric(citibike_may23_df['end_station_id'], errors='coerce')
citibike_may24_df.loc[:, 'start_station_id'] = pd.to_numeric(citibike_may24_df['start_station_id'], errors='coerce')
citibike_may24_df.loc[:, 'end_station_id'] = pd.to_numeric(citibike_may24_df['end_station_id'], errors='coerce')

citibike_may23_df = citibike_may23_df.dropna()
citibike_may24_df = citibike_may24_df.dropna()

# Convert the start_station_id and end_station_id columns to float
citibike_may23_df['start_station_id'] = citibike_may23_df['start_station_id'].astype('float64')
citibike_may23_df['end_station_id'] = citibike_may23_df['end_station_id'].astype('float64')
citibike_may24_df['start_station_id'] = citibike_may24_df['start_station_id'].astype('float64')
citibike_may24_df['end_station_id'] = citibike_may24_df['end_station_id'].astype('float64')

citibike_may23_df.info()
citibike_may24_df.info()
```

```python
# Combine the dataframes
citibike_may2324_df = pd.concat([citibike_may23_df, citibike_may24_df])

# Reset the index
citibike_may2324_df.reset_index(drop=True, inplace=True)

citibike_may2324_df.head()
```

```python
citibike_may2324_df.info()
```

```python
# Write to CSV
citibike_may2324_df.to_csv('Resources/202405-202305-citibike-tripdata.csv', index=False)
```
